```python
import pandas as pd

df = pd.read_csv('customer_shopping_behaviour.csv') 
```

```python
df.isnull().sum()
#there was 37 null values before#
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
#by doing this it will not cake the overall median(avg) of the entire data but the median value of the data depending on its category#
```

```python
df.isnull().sum()
#now its 0#
```

```python
df.columns
```

```python
df.columns = df.columns.str.lower()#will turn everything into lowercase#
df.columns = df.columns.str.replace(' ','_')#will replace every space to underscore#
```

```python
df.columns
```

```python
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})#that usd part is not required#
```

```python
df.columns #now all the values are in snake case successfully #
```

```python
# create a column age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age', 'age_group']].head(15)
```

```python
# create column purchase_frequency_days
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months':90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(15)
```
